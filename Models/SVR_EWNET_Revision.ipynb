## SVR

```python
# from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
from sktime.forecasting.model_selection import temporal_train_test_split
from sktime.utils.plotting import plot_series
from sklearn.svm import SVR
from sktime.forecasting.compose import make_reduction
```

## Weekly data

```python
# import data
data = pd.read_csv('/content/Iquitos_data_weekly.csv')
data.head()
```

```python
#We only take cases because we want univariate data
y = data['Cases']
type(y)
```

### Long Term

```python
n = 52
y_train, y_test = y[:-n], y[-n:]
fh = np.arange(1, n+1)
plot_series(y_train, y_test, labels=["y_train", "y_test"]);
```

```python
# declarative forecaster specification - just two lines!
regressor = SVR(C=1.0, epsilon=0.2)
forecaster = make_reduction(regressor, window_length=15, strategy="recursive")
forecaster.fit(y_train)
pred = forecaster.predict(fh)
```

```python
from google.colab import files
pred.to_csv('Iqt_den_52_SVR.csv', encoding = 'utf-8-sig') 
files.download('Iqt_den_52_SVR.csv')
```
